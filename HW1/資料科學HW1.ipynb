```python
def data_to_np(data):                                       #將dataframe格式資料以np array做轉換，方便後續的計算與預測
    input=[]
    for i in range(len(data)):
        x = data.values[i]
        input.append(x)
    data = [np.array(datas)for datas in input]
    data = np.array(data)
    return data
def computedistance(v1,v2):                                 #此function是將data當作是一個vector，並且去計算兩個data之間的距離
    distance = 0.0
    for i in range (len(v1)):
        distance += pow((v1[i]-v2[i]),2)
    distance = math.sqrt(distance)
    return distance
def getneighbor(v1,v2,choose):                              #選取離這個data最近距離的data set，choose是選取幾個最近的data
    distance = 0.0
    sort_distance = []
    neighbor = []
    for i in range(len(v1)):
        distance = computedistance(v1[i][:-1],v2)
        sort_distance.append((v1[i],distance))
    sort_distance.sort(key = operator.itemgetter(1))
    for i in range(choose):
        neighbor.append(sort_distance[i][0])
    return neighbor
def getresult(neighbor):                                    #將離最近的data set去決定yes/no的多寡，yes多則去結果為yes，反之則反
    yes= 0
    no = 0
    for i in range(len(neighbor)):
        if(neighbor[i][-1]==0):
            no+=1
        else:
            yes+=1
    if(no>=yes):
        return 0
    else:
        return 1
```

```python
temp = process_data(input,0)
training_data = data_to_np(temp)
```

```python
output = pd.read_csv("testing_data.csv")
temp = process_data(output,1)
testing_data = data_to_np(temp)
```

```python
answer = []
for i in range(len(testing_data)):
    neighbor = []
    neighbor = getneighbor(training_data,testing_data[i],5)
    answer.append(getresult(neighbor))
answer
```
